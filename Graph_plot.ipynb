```python
#plot function y=x**3
import numpy as np
x=np.array([1,4,5])
y=x**3
plt.plot(x,y,"r--")
plt.show()
```

```python
#make the above graph smooth
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
#plt.plot(x,y,color="black",marker=".")
plt.plot(x,y,color="black",linewidth=5,label="x^3")
plt.plot(x,y2,color="red",linewidth=5,label="x^2")
#to make label working we have to include following line
plt.legend()
plt.ylabel("y")
plt.xlabel("x")
plt.grid()
#if you want to have different axis
#plt.axis(x_start_ind,x_end_ind,y_start_ind,y_end_ind)
plt.axis([0,10,0,100])
plt.title("matplotlibe_demo")
# add random text at some position
plt.text(2,80,"test",fontsize=20)
plt.show()
```

```python
#for single variable x will be 0 to range(size of array)
a=[4,5,6]
plt.plot(a)
plt.show()
```

```python
# pie graph
labels=["A","B","C","D"]
sizes=[3,4,6,2]
color=["red","green","blue","yellow"]
explode=[0.1,0.2,0,0]
plt.pie(sizes,colors=color,labels=labels,autopct="%.2f%%",explode=explode,counterclock=False,startangle=90) #the axix will not be equal
plt.title("split among classes")
plt.axis("equal")

plt.show()
```
